```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('zzz.csv')
```

```python
df.value_counts('6')
```

```python
df['6'].describe()
```

```python
# 创建箱形图
fig, ax = plt.subplots()
ax.boxplot(df['6'])

# 添加标题和标签
ax.set_title('IMDB Movie Ratings')
ax.set_ylabel('Rating')

# 显示图表
plt.show()
```

```python
counts = df['6'].value_counts()

# 创建条形图
plt.bar(counts.index, counts)

# 添加标题和标签
plt.title('IMDB Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')

# 显示图表
plt.show()
```

```python
bins = [-0.1, 4, 7, 9.1]

labels = [0,1,2]

df['8'] = pd.cut(df['6'], bins=bins, labels=labels)
```

```python
bins = [-0.1, 1,3, 5, 7, 10.1]

labels = [0, 1, 2, 3, 4]

df['8'] = pd.cut(df['6'], bins=bins, labels=labels)
```

```python
df['8'].value_counts()
```
